Come ci si aspetterebbe il valore di $\pi$ tende a quello che si sa essere il valore esatto e le barre di errore diminuiscono. 
Per fare l'esperimento di buffon ho costruito:
- Un quadrato di $l = 10000$
- $d = 2$
- $l = 1.5$

Poi ho estratto casualmente un punto nel quadrato e un angolo per costruirmi la posizione del bastoncino e vedere se intersecava le linee del piano. Ovviamente la probabilità è stata ottenuta dividente il numero di incroci e il numero di tentativi totali.

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math

file = open("misurePI.txt", "r")

for line in file:
    i=line.split()
    plt.errorbar(float(i[0]),float(i[1]), yerr=float(i[2]),marker='.', color='steelblue')

plt.xlabel('# throws')
plt.ylabel('Pi')
plt.title('Misure di Pi')
plt.grid(True)
plt.show()
```
