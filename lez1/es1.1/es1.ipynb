# Es 1.1-1.2

Come da consegna sono stati svolti i 2 integrali assegnati con metodo media a blocchi
- $\langle r \rangle = \int_0^1 r dr = 1/2$
- $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math

N=100
L=1000
x = np.arange(N)      # [0,1,2,...,N-1]
sum_prog = np.zeros(N)
err_prog = np.zeros(N)

file = open("sum_prog1.txt", "r")
i=0
for line in file:
    sum_prog[i]=float(line)
    #print(sum_prog[i])
    i=i+1
file.close()

file = open("err_prog1.txt", "r")
i=0
for line in file:
    err_prog[i]=float(line)
    #print(sum_prog[i])
    i=i+1
file.close()

x*=L
plt.errorbar(x,sum_prog-0.5,err_prog)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()


file = open("sum_prog2.txt", "r")
i=0
for line in file:
    sum_prog[i]=float(line)
    #print(sum_prog[i])
    i=i+1
file.close()

file = open("err_prog2.txt", "r")
i=0
for line in file:
    err_prog[i]=float(line)
    #print(sum_prog[i])
    i=i+1
file.close()

x*=L
plt.errorbar(x,sum_prog-1/12,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()

```
