## 32 citt√† randomiche

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math


x,y = np.loadtxt("cities.txt", usecols=(0,1), delimiter=" ", unpack='true')
plt.plot(x,y,linestyle="",marker='o')
plt.ylabel(f'y')
plt.xlabel(f'x')
plt.show()
```

```python
y = np.loadtxt("path.txt", usecols=(0), delimiter=" ", unpack='true')
plt.ylabel(f'$|L|^2$')
plt.xlabel(f'generations')
plt.plot(y,marker='.')
plt.figure()
y = np.loadtxt("temp.txt", usecols=(0), delimiter=" ", unpack='true')
plt.ylabel(f'$T$')
plt.xlabel(f'generations')
plt.plot(y,linestyle="",marker='.')
plt.show()
```

#### Miglior percorso

```python
x,y = np.loadtxt("bestpath.txt", usecols=(0,1), delimiter=" ", unpack='true')
width1 = 10
height1 = 10
width_height_1 = (width1, height1)
plt.figure(figsize=width_height_1)
plt.plot(x,y,linestyle="dashed",marker='o')
plt.ylabel(f'y')
plt.xlabel(f'x')
plt.show()
```

#### Rating di accettazione del metropolis

```python
x,y,z = np.loadtxt("metropolisAccTot.txt", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.plot(z,x/y,linestyle="",marker='o')
plt.ylabel(f'Acc/Tot')
plt.xlabel(f'T')
plt.show()
```
