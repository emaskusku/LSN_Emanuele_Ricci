## 32 città sulla circonferenza 

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math


x,y = np.loadtxt("cities.txt", usecols=(0,1), delimiter=" ", unpack='true')
plt.plot(x,y,linestyle="",marker='o')
plt.ylabel(f'y')
plt.xlabel(f'x')
plt.show()
```

#### Lunghezza del percorso più breve per ogni generazione

```python
y = np.loadtxt("bestpathgen.txt", usecols=(0), delimiter=" ", unpack='true')
plt.ylabel(f'$|L|^2$')
plt.xlabel(f'generations')
plt.plot(y,linestyle="",marker='.')
plt.show()
```

#### Media della metà delle lunghezze più corte della generazione

```python
y = np.loadtxt("avebestpath.txt", usecols=(0), delimiter=" ", unpack='true')
plt.ylabel(f'$<|L|>^2$')
plt.xlabel(f'generations')
plt.plot(y,linestyle="",marker='.')
plt.show()
```

#### Percorso ottimale trovato

```python
x,y = np.loadtxt("bestpath.txt", usecols=(0,1), delimiter=" ", unpack='true')
width1 = 10
height1 = 10
width_height_1 = (width1, height1)
plt.figure(figsize=width_height_1)
plt.plot(x,y,linestyle="dashed",marker='o')
plt.ylabel(f'y')
plt.xlabel(f'x')
plt.show()
```
