```python
import numpy as np
import matplotlib.pyplot as plt

ave_g_solid = np.loadtxt(fname = "output.gave.solid.0",usecols=0)
#err_g_liquid = np.loadtxt(fname = "output.gave_Liquid.0",usecols=1)

npart=108.0
rho=0.8
vol=npart/rho
box=vol**(1./3.)
L=box/2.
bins = np.linspace(start=0, stop=L, num=100)

plt.plot(bins,ave_g_solid,color='steelblue')

plt.xlabel('r')
plt.ylabel('g(r)')
plt.grid(True)
plt.show()
```
