# Es 7.3-7.4

Segue un confronto tra il grafico della g(r) ottenuta con il codice di dinamica molecolare modificato e quello con il Monte Carlo MVT.
I risultati paiono in accordo come ci si aspetterebbe 

### Liquid

```python
import numpy as np
import matplotlib.pyplot as plt

ave_g_liquid_NVT = np.loadtxt(fname = "es7.2/output.gave_Liquid.0",usecols=0)
err_g_liquid_NVT = np.loadtxt(fname = "es7.2/output.gave_Liquid.0",usecols=1)
ave_g_liquid_MD = np.loadtxt(fname = "es7.3/output.gave_Liquid.0",usecols=0)
err_g_liquid_MD = np.loadtxt(fname = "es7.3/output.gave_Liquid.0",usecols=1)

npart=108.0
rho=0.8
vol=npart/rho
box=vol**(1./3.)
L=box/2.
bins = np.linspace(start=0, stop=L, num=100)

plt.errorbar(bins,ave_g_liquid_NVT,yerr=err_g_liquid_NVT, lw=1 ,color='red', elinewidth=5,label='MC_NVT')
plt.errorbar(bins,ave_g_liquid_MD,yerr=err_g_liquid_MD, lw=1 ,color='blue', elinewidth=5,label='MD')
plt.legend()

plt.xlabel('r')
plt.ylabel('g(r)')
plt.grid(True)
plt.show()
```
