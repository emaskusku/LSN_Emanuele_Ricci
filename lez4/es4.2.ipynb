### Energia cinetica

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math

x,y,error = np.loadtxt("MolecularDynamics_NVE/ave_ekin.out", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(x,y,yerr=error)
plt.xlabel('#values')
plt.ylabel('Ekin [j]')
plt.grid(True)
```
