# Es 3

Di seguito sono riportati i grafici di argon allo stato solido, liquido e gas. Ogni volta è stata eseguita una equilibrazione per poi procedere alla misurazione.
Le unità di misura sono state tenute ridotte perchè i fattori di conversione erano esorbitanti e anche inserendoli non aumentavano la comprensione del problema. 
Solo la temperatura è stata espressa in kelvin, proprio perchè bastava moltiplicarla per 120. 
In ogni caso sotto sono riportate tutti le grandezze necessarie per convertire le unità ridotte in quelle del SI

```python
epsilon_kB = 120. #kelvin
kb = 1.380649*(10.**(-23))
epsilon = 120*kb
sigma=0.34*(10**(-9))
```
