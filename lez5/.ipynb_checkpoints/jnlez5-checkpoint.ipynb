# Es 1
Di seguito è riportato il grafico della media a blocchi con 100 blocchi per la per $\Psi$$_{1,0,0}$ e $\Psi$$_{2,1,0}$ usando una transition probability uniforme dell'algoritmo di metropolis e successivamente una gaussiana.
Per capire la dimensione dei blocchi ho usato un metodo euristico: partendo dalla posizione iniziale di (1,0,0) ho contato quanti step occorrevano per far diventare negativa la coordinata x, poi ho moltiplicato per 10 tale valore. Una misurazione più precisa sarebbe stata ottenuta usando l'autocorrelazione (cosa che ho fatto nell'esercitazione successiva). 
Ho mantenuto il valore di accettazione del metropolis al 50% per evitare il più possibile misurazioni correlate ed in entrambi i casi i risultati sono stati soddisfacenti.
Ho notato che partendo da una posizione eccessivamente lontana il metropolis non riesce a performare adeguatamente poichè non riesce ad arrivare in zone dove la probabilità aumenta.

### Transition probability uniforme

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math

x,y,error = np.loadtxt("psi1.txt", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(x,y,yerr=error)
plt.title("$\Psi_{1,0,0}$")
plt.xlabel('#throws')
plt.ylabel('raggio medio [$a_0$]')
plt.grid(True)
```

```python
x,y,error = np.loadtxt("psi2.txt", usecols=(0,1,2), delimiter=" ", unpack='true')
plt.errorbar(x,y,yerr=error)
plt.title("$\Psi$$_{2,1,0}$")
plt.xlabel('#throws')
plt.ylabel('raggio medio [$a_0$]')
plt.grid(True)
```
